```python
import pandas as pd

# Load your current dataset
url = "https://raw.githubusercontent.com/aaut0001/FIT3179_W10/main/airline.csv"
df = pd.read_csv(url)

# Convert the 'Departure Date' column to datetime format
df['Departure Date'] = pd.to_datetime(df['Departure Date'])

# Create a new column 'Month Name' based on 'Departure Date'
df['Month Name'] = df['Departure Date'].dt.strftime('%B')

# Save the modified dataset to a new CSV file
df.to_csv("modified_airline.csv", index=False)
```

```python
import pandas as pd

# Load the sample dataset (replace 'sample_dataset.csv' with your actual dataset URL or file path)
data = pd.read_csv('modified_airline.csv')

# Group the data by 'Continents' and calculate the total flights for each country
grouped = data.groupby(['Continents', 'Country Name']).size().reset_index(name='Total Flights')

# Rank the countries within each continent based on total flights
grouped['Rank'] = grouped.groupby('Continents')['Total Flights'].rank(ascending=False, method='min')

# Filter the top 3 countries for each continent
top_countries = grouped[grouped['Rank'] <= 3]

# Create a new column 'Country Name (Total Flights)' for tooltips
top_countries['Country Name (Total Flights)'] = top_countries['Country Name'] + ' (' + top_countries['Total Flights'].astype(str) + ')'

# Save the new dataset as a CSV file
top_countries.to_csv('top_countries_dataset.csv', index=False)
```
